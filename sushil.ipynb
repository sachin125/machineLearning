```r
house_prices <- read.csv("/content/train.csv", header = TRUE)
```

```r
head(house_prices)
```

Histogram

```r
hist(house_prices$LotArea, breaks = 30, xlab = "Lot Area (square feet)", main = "Distribution of
Lot Area")
```

```r
hist(house_prices$GrLivArea, breaks = 30, xlab = "Above Ground Living Area (square feet)", main
= "Distribution of Above Ground Living Area")
```

```r
hist(house_prices$BedroomAbvGr, breaks = 5, xlab = "Number of Bedrooms Above Basement
Level", main = "Distribution of Bedrooms Above Basement Level")
```
